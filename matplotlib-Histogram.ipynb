# Histogram 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use("fivethirtyeight")
```

## Histogram are used to represent the continuous distribution of values over multiple bins

```python
ages = [18,19,21,25,26,30,32,38,45,55]

plt.hist(ages)

plt.title('Ages of students')
plt.xlabel('Ages')
plt.ylabel('Total Student')
```

## ploting the graph in 5 bins only

```python
ages = [18,19,21,25,26,30,32,38,45,55]

plt.hist(ages, bins = 5)

plt.title('Ages of students')
plt.xlabel('Ages')
plt.ylabel('Total Student')
```

## setting edge color to the bins

```python
ages = [18,19,21,25,26,30,32,38,45,55]
bins = [10,20,30,40,50,60]

plt.hist(ages, bins = bins ,edgecolor = "black", )

plt.title('Ages of students')
plt.xlabel('Ages')
plt.ylabel('Total Student')
```

## excluding one range of student from the plot

```python
ages = [18,19,21,25,26,30,32,38,45,55]
bins = [20,30,40,50,60]

plt.hist(ages, bins = bins ,edgecolor = "black")

plt.title('Ages of students')
plt.xlabel('Ages')
plt.ylabel('Total Student')
```

## showing median age on the plot

```python
ages = [18,19,21,25,26,30,32,38,45,55]
bins = [20,30,40,50,60]

median_age = 27

plt.hist(ages, bins = bins ,edgecolor = "black", )

plt.axvline(median_age, color = 'red', label = 'Median Age')

plt.title('Ages of students')
plt.xlabel('Ages')
plt.ylabel('Total Student')
plt.legend()
```

```python
ages = [18,19,21,25,26,30,32,38,45,55]
bins = [10,20,30,40,50,60]

median_age = 27

plt.hist(ages, bins = bins ,edgecolor = "black", )

plt.axvline(median_age, color = 'red', label = 'Median Age',linewidth = 2)

plt.title('Ages of students')
plt.xlabel('Ages')
plt.ylabel('Total Student')
plt.legend()
```
