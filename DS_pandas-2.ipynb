## Let's start with dictionary as Container

```python
person = {
    "first" : "priyanshu",
    "last"  : "verma",
    "email" : "priyanshuverma@gmail.com"
}
```

```python
person
```

### But here the list can store only one value for each key
### lets go for list of values for each key

```python
people = {
    "first" : ["priyanshu","anirudh","naresh"],
    "last" : ["verma","kushwah","kumar"],
    "email" : ["priyanshukumarv984@gmail.com","anirudh@gmail.com","naresh@gmail.com"]
}
```

```python
people
```

```python
#accessing specific value
people['email']
```

#### problem with disctionary is: it has very limited functionalities
### hence lets go to pandas and it's data frame

## DataFrame : 2-D data structure (Rows and column)


```python
df = pd.DataFrame(people)
```

```python
df
```

```python
#accessing single value
df["email"]
```

### the above printed object is a series
## Series : 1-D array (rows of single column)
# DataFrame : is a container of multipleseries

```python
type(df['email'])
```

```python
df.email
```

```python
# acceessing multiple column
df[["first","last"]]
```

```python
df
```

```python
# knowing all the column of data frame
df.columns
```

## To access rows loc and iloc are used
### loc : it is lable-based
### iloc : it is integer positioned-based

```python
# To acess single row
df.iloc[0]
```

```python
# To accessing multiple rows
df.iloc[[0,1]]
```

```python
#rows of specific column only
df.iloc[[0,1],2]
```

```python
# multiple rows and multiple column
df.iloc[[0,1],[0,1]]
```

```python
df.loc[0]
```

```python
df.loc[[0,1],["email","first"]]
```
