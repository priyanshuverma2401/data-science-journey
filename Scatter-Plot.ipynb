# scstter-plot
## used when relationship between 2 sets of values to be compared and to see if they are co-related

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn-v0_8')
```

```python
x = [5,7,8,5,6,7,9,2,3,4,4,4,2,6,3,6,8,6,4,1]
y = [7,4,3,9,1,3,2,5,2,4,8,7,1,6,4,9,7,7,4,1]
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y, s=100)
```

```python
plt.scatter(x,y, s=100, c = "green")
```

```python
plt.scatter(x,y, s=100, c = "green" , marker = 'x')
```

```python
plt.scatter(x,y, s=100, c = "green", linewidth = 3,edgecolor = 'black')
```

```python
plt.scatter(x,y, s=100, c = "green", linewidth = 3,edgecolor = 'black', alpha = 0.75)
```

## adding additional infoemation to dataset using color

```python
import pandas as pd
```

```python
colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
plt.scatter(x,y, s=100, c = colors, edgecolor = "black", linewidth = 1, alpha = 0.75)
```

```python
colors = [7,5,9,7,5,7,2,5,3,7,1,2,8,1,9,2,5,6,7,5]
plt.scatter(x,y, s=100, c = colors, edgecolor = "black", cmap = 'Greens',linewidth = 1, alpha = 0.75 )
cbar = plt.colorbar()
cbar.set_label('Rating')
```

```python
data = pd.read_csv("insurance.csv")
```

```python
data
```

```python
age = data['age']
bmi = data['bmi']
charges = data['charges']
```

```python
plt.xlabel("age")
plt.ylabel('bmi')

plt.scatter(age, bmi, s = 100, edgecolor = 'black', c = charges, cmap = 'summer', linewidth =2, alpha = 0.75)
cbar = plt.colorbar()
cbar.set_label('Rating')
```

```python
plt.xlabel("age")
plt.ylabel('bmi')

plt.scatter(age, bmi, s = 100, edgecolor = 'black', c = charges, cmap = 'summer', linewidth =2, alpha = 0.75)
cbar = plt.colorbar()
cbar.set_label('Rating')
plt.xscale('log')
plt.yscale('log')
```
