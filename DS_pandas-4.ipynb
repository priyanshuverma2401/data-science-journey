# Filtering

```python
import pandas as pd
```

```python
people = {
    "first" : ["priyanshu","anirudh","naresh"],
    "last" : ["verma","kumar","kumar"],
    "email" : ["priyanshukumar984@gmail.com","anirudh@gmail.com","naresh@gmail.com"]
}
```

```python
df = pd.DataFrame(people)
df
```

```python
df['last'] == 'kumar'
```

```python
df.loc[df['last'] == 'kumar']
```

```python
filt = (df['last'] == 'kumar')
```

```python
df.loc[filt]
```

```python
df.loc[filt, 'email']
```

```python
df.loc[filt,['first','email']]
```

# AND and OR in pandas filtering
# AND = &
# OR = |

```python
#and
filt = ((df['last'] == 'kumar') & (df['first']== 'naresh'))
```

```python
df.loc[filt]
```

```python
# OR
filt = ((df['last']=='kumar') | (df['first']=='priyanshu'))
```

```python
df.loc[filt]
```

## NOT = ~

```python
df.loc[~filt]
```
